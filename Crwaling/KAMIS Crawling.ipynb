## KAMIS Crawling

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# 원래는 100일 자료제한 (기상청 평균가격 크롤링 - 1개년)
# 날짜랑 가격을 같이 가져와야한다.

date_list = [('2019-01-02', '2019-06-30'),
            ('2018-01-02', '2018-06-30'),
            ('2018-07-02', '2018-12-31'),
            ('2017-01-02', '2017-06-30'),
            ('2017-07-02', '2017-12-31'),
            ('2016-01-02', '2016-06-30'),
            ('2016-07-02', '2016-12-31'),
            ('2015-01-02', '2015-06-30'),
            ('2015-07-02', '2015-12-31'),
            ('2014-01-02', '2014-06-30'),
            ('2014-07-02', '2014-12-31'),
            ('2013-01-02', '2013-06-30'),
            ('2013-07-02', '2013-12-31'),
            ('2012-01-02', '2012-06-30'),
            ('2012-07-02', '2012-12-31'),
            ('2011-01-02', '2011-06-30'),
            ('2011-07-02', '2011-12-31'),
            ('2010-01-02', '2010-06-30'),
            ('2010-07-02', '2010-12-31'),
            ('2009-01-02', '2009-06-30'),
            ('2009-07-02', '2009-12-31'),
            ('2008-01-02', '2008-06-30'),
            ('2008-07-02', '2008-12-31'),]

#print(date_list[0][0])



def get_day_price(index):
    global year_data
    url = 'https://www.kamis.or.kr/customer/price/retail/period.do?action=daily&startday='+str(date_list[index][0])+'&endday='+str(date_list[index][1])+'&countycode=&itemcategorycode=200&itemcode=232&kindcode=&productrankcode=&convert_kg_yn=N'
    soup = BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")

    '''
    ===============================
                  날짜
    ===============================
    '''
    # 년도 가져오기
    year_data = int(date_list[index][0][:4])
    
    # 날짜 가져오기 
    days = []
    for link in soup.find_all('th', class_=''):
        w = link.text
        if len(w) == 5: # 날짜 형태인것만 가져온다.
            add_text = str(year_data) + '/' + w  # 년도와 날짜 합치기 
            days.append(add_text)
            
    '''
    ===============================
                  가격
    ===============================
    '''
    daily_prices = []
    fwb_list = soup.find_all('tr', class_="fwb") # 'fwb' 하위 정보들이 배열로 담긴다.
    
    td_list = fwb_list[0].find_all('td', class_='') # 'td구간에 해당하는' 모든 글자들이 배열로 담긴다.
    for i in range(len(td_list)):
        price = td_list[i].text    # 해당td의 '글자'가 담긴다 (text) :  <td class=""> 이 부분 </td>
        daily_prices.append(price)
        
    return days[:len(days)//2], daily_prices





index = 0

while index<22:
    days,daily_prices = get_day_price(index)
    #print(days)
    print("총",len(days),"개의 날짜를 찾았습니다.")
    #print(daily_prices)
    print("총",len(daily_prices),"개의 가격을 찾았습니다.")

    '''
    ===============================
                크롤링
    ===============================
    '''
    # json 형태로 변환
    res = {}	
    c = ['date', 'price']
    res[c[0]] = days
    res[c[1]] = daily_prices

    #print(res)

    # pandas를 이용해 엑셀로 쓰기
    try:
        df = pd.DataFrame(res, columns = c) # 크롤링 결과물이 pandas 가 읽을 수 있게 변환되어 data 라는 변수로 들어간다.
        file_name = 'test+['+str(year_data)+'-'+ str(index%2)+'].csv'
        df.to_csv(file_name,encoding="cp949") # 엑셀 파일로 저장
        print(file_name,': 저장 완료!')
    except Exception as e:
        print('저장 오류 ! ', e)
        
    index += 1
```
